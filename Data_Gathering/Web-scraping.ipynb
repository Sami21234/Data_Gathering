## In this notebook, we will scrap the data from the Ambitionbox.com which shows the company, their ratings and others, etc 

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
webPage = requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=1', headers=headers).text
```

```python
soup = BeautifulSoup(webPage, 'lxml')
```

```python
# print(soup.prettify())      # prettify() --> this will write the html code in the pretty manner

soup.find_all('h1')[0].text.strip()
```

## Now, we will scrap the entire block1, which have all the informartion like rating, reviews, place, salary, etc

```python
company_data = soup.find_all('div', class_= 'companyCardWrapper')
len(company_data)
```

```python
company_name = []
ratings = []
reviews = []
company_type = []
locations = []

for i in company_data:
    company_name.append(i.find('h2').text.strip())
    ratings.append(i.find('div', class_= 'rating_text').text.strip())
    reviews.append(i.find('span', class_= 'companyCardWrapper__companyRatingCount').text.strip())
    company_type.append(i.find_all('span', class_= 'companyCardWrapper__interLinking')[0].text.split('|')[0].strip())
    locations.append(i.find_all('span', class_= 'companyCardWrapper__interLinking')[0].text.split('|')[-1].strip())

d = {'company_name': company_name, 'rating': ratings, 'reviews': reviews, 'company_type': company_type, 'locations': locations}
df = pd.DataFrame(d)
df
```

## Now, for all the pages

```python
final_data = pd.DataFrame()
for j in range(1, 300):

    url = 'https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(j)

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
    webPage = requests.get(url, headers=headers).text

    soup = BeautifulSoup(webPage, 'lxml')
    company_data = soup.find_all('div', class_= 'companyCardWrapper')

    company_name = []
    ratings = []
    reviews = []
    company_type = []
    locations = []

    for i in company_data:
        company_name.append(i.find('h2').text.strip())
        ratings.append(i.find('div', class_= 'rating_text').text.strip())
        reviews.append(i.find('span', class_= 'companyCardWrapper__companyRatingCount').text.strip())
        company_type.append(i.find_all('span', class_= 'companyCardWrapper__interLinking')[0].text.split('|')[0].strip())
        locations.append(i.find_all('span', class_= 'companyCardWrapper__interLinking')[0].text.split('|')[-1].strip())

    d = {'company_name': company_name, 'rating': ratings, 'reviews': reviews, 'company_type': company_type, 'locations': locations}
    df = pd.DataFrame(d)

    final_data=pd.concat([final_data, df], ignore_index=True)
    
```
